```python
import fiftyone as fo

# Create a new dataset
dataset = fo.Dataset()

# Create a sample
sample = fo.Sample(filepath="./coco_dataset/data/000000001393.jpg")

# Create a detection (bounding box + label)
detection = fo.Detection(label="cat", bounding_box=[0.1, 0.2, 0.5, 0.6])

# Add the detection to the sample
sample["ground_truth"] = fo.Detections(detections=[detection])

# Add the sample to the dataset
dataset.add_sample(sample)

# Launch the app to visualize the dataset
session = fo.launch_app(dataset)
```
